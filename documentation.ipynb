# MODULE 3: GLCM FEATURES 

<img src='glcmm.gif'>

There are four GLCM features,<br><br>
+ ```correlation()```
+ ```homogeneity()```
+ ```contrast()```
+ ```energy()```,
and to extract all of these features at ones the package has a,
**_```glcm__all()```_** function.





<h2><center>1. Correlation</center></h2>

Given an image matrix **(numpy.ndarray)** the function **```glcm(image-array).correlation()```** returns the correlation value of the _image/numpy.ndarray_

````python
glcm(n_array[1]).correlation()
````

````python
glcm(n_array[1]).glcm_all()
````

<h2><center>2. Homogeneity</center></h2>

Given an image matrix **(numpy.ndarray)** the function **```glcm(image-array).homogeneity()```** returns the homogeneity values of the _image/numpy.ndarray_

````python
glcm(n_array[1]).homogeneity()
````

<h2><center>3. Contrast</center></h2>

Given an image matrix **(numpy.ndarray)** the function **```glcm(image-array).contrast()```** returns the contrast values of the _image/numpy.ndarray_

````python
glcm(n_array[1]).contrast()
````

<h2><center>4. Energy</center></h2>

Given an image matrix **(numpy.ndarray)** the function **```glcm(image-array).energy()```** returns the energy values of the _image/numpy.ndarray_

````python
glcm(n_array[1]).energy()
````

<h2><center>5. Every GLCM Feature</center></h2>

Given an image matrix **(numpy.ndarray)** the function **```glcm(image-array).glcm_all()```** returns the every feature value of the _image/numpy.ndarray_

````python
glcm(n_array[1]).glcm_all()
````

# MODULE 4: MOMENTS

There are two types of Moments,
1. Scale Variant which changes with the change in scale, transformation and rotation, and
    - 24 values
2. Scale Invariant moments or Hu moments which aren't affected by the transformations in scale or any rotation. 
    - 7 values

```get_moments()``` - returns the 24 variant moments

````python
from xtract_features.moments import *
````

````python
moments(n_array[1]).get_moments()
````

```get_HuMoments()``` - returns 7 invariant Hu Moments as proposed by Hu.

````python
moments(n_array[1]).get_HuMoments()
````
